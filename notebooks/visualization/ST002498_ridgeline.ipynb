```python

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import joypy

output_directory = '/home/marco/DataspellProjects/thesis/output/visualization/'

palette = sns.color_palette(["#2176AB", "#F97662", "#FFBF00", "#50C878", "#B284BE"])
sns.set(style="whitegrid", font_scale=1.5)
sns.set_palette(palette)

df = pd.read_csv('../../data/archive/ST002498.csv')
df_normalized = pd.read_csv('../../data/archive/ST002498_balanced_standard.csv')
df_unit = pd.read_csv('../../data/archive/ST002498_balanced_unit.csv')

```

```python
def ridgeline_plot(data, title, color):
    fig, axes = joypy.joyplot(data.iloc[:, 90:140], alpha=0.25,
                              linewidth=0.3, color=color, linecolor=color, figsize=(8, 6), x_range=[-1, 1.5],
                              ylabels=False)
    
    plt.suptitle(title, fontsize=16)
    plt.savefig(output_directory + 'ridgeline/' + title + '.png', dpi=300)

#ridgeline_plot(df, 'Original', palette[0])
#ridgeline_plot(df_normalized, 'Normalized', palette[1])
ridgeline_plot(df_unit, 'Unit', palette[3])
```
