```python
import pandas as pd

# Leggi il file CSV
df = pd.read_excel("../data/ST000369.xlsx")  # Sostituisci "percorso_del_tuo_file.csv" con il percorso effettivo del tuo file CSV

# Seleziona solo le colonne numeriche
numeric_cols = df.select_dtypes(include=['float64', 'int64'])
cov_matrix = numeric_cols.cov()

# Funzione personalizzata per evidenziare valori superiori a una certa soglia
def highlight_high_values(val, threshold=5):
    color = 'yellow' if abs(val) > threshold else 'black'
    return f'color: {color}'

styled_cov_matrix = cov_matrix.style.background_gradient(cmap='coolwarm', axis=None) \
    .applymap(highlight_high_values)

styled_cov_matrix
```
